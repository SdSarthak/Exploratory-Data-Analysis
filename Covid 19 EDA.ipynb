# **Import Libraries and Dataset**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
data = pd.read_csv(url)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
# Check for missing values before cleaning
print("\nMissing values before cleaning:")
print(data.isnull().sum().sort_values(ascending=False))
```

```python
# Check for duplicates
duplicates = data.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
data_cleaned = data.dropna(axis=1, how='any')
#Remove them
if duplicates > 0:
    data = data.drop_duplicates()
    print(f"Duplicates removed. New number of rows: {data.shape[0]}")
```

```python
# Select essential columns to keep
essential_columns = ['date', 'location', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths']
data_essential = data[essential_columns]
```

```python
# Drop rows with NaN in essential columns only
data_cleaned = data_essential.dropna(subset=essential_columns)
```

```python
data_cleaned.shape
```

```python
data_cleaned.isnull().sum().sort_values(ascending=False)
```

```python
# Convert 'date' column to datetime format if it exists
if 'date' in data_cleaned.columns:
    data_cleaned['date'] = pd.to_datetime(data_cleaned['date'])
```

```python
# Reset the index of the cleaned DataFrame
data_cleaned.reset_index(drop=True, inplace=True)
```

```python
data_cleaned.describe()
```

# **Visualisation**


**Distribution of total cases per country**

```python
plt.figure(figsize=(12, 6))
top_countries = data_cleaned.groupby('location')['total_cases'].max().nlargest(10).index
sns.barplot(data=data_cleaned[data_cleaned['location'].isin(top_countries)],
             x='total_cases',
             y='location',
             estimator=max,
             palette='viridis')
plt.title('Top 10 Countries by Total COVID-19 Cases')
plt.xlabel('Total Cases')
plt.ylabel('Country')
plt.show()
```

**The trend of cases over time for a specific country**

```python
usa_data = data_cleaned[data_cleaned['location'] == 'United States']
plt.figure(figsize=(14, 7))
sns.lineplot(data=usa_data, x='date', y='total_cases', marker='o', label='Total Cases')
sns.lineplot(data=usa_data, x='date', y='total_deaths', marker='o', label='Total Deaths', color='red')
plt.title('COVID-19 Cases and Deaths in the USA Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Correlation heatmap for numerical features
plt.figure(figsize=(10, 8))

# Select only numerical columns for correlation
numerical_data = data_cleaned.select_dtypes(include=['float64', 'int64'])
if not numerical_data.empty:
    correlation = numerical_data.corr()
    sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap of Numerical Features')
else:
    print("No numerical data available for correlation analysis.")

plt.tight_layout()
plt.show()
```

```python
# Aggregate total cases and total deaths by date for visualization
conclusion_data = data_cleaned.groupby('date').agg({
    'total_cases': 'sum',
    'total_deaths': 'sum'
}).reset_index()
```

#**Conclusion**

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=conclusion_data, x='date', y='total_cases', marker='*', label='Total Cases', color='blue')
sns.lineplot(data=conclusion_data, x='date', y='total_deaths', marker='*', label='Total Deaths', color='red')
plt.title('Total COVID-19 Cases and Deaths Over Time Globally')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```
