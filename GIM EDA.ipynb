```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('GIM.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
data['Physical Area (ha) 2016-17'] = pd.to_numeric(data['Physical Area (ha) 2016-17'], errors='coerce')
data['Physical Area (ha) 2017-18'] = pd.to_numeric(data['Physical Area (ha) 2017-18'], errors='coerce')
```

```python
data.isnull().sum()
```

```python
data.fillna(method='ffill', inplace=True)
```

```python
data.dropna(thresh=3, inplace=True)
```

```python
data['Released Amount (In Rs. Crore) 2015-16'].fillna(data['Released Amount (In Rs. Crore) 2015-16'].mean(), inplace=True)
data['Physical Area (ha) 2015-16'].fillna(data['Physical Area (ha) 2015-16'].mean(), inplace=True)

data['Released Amount (In Rs. Crore) 2019-20'].fillna(data['Released Amount (In Rs. Crore) 2019-20'].mean(), inplace=True)

data['Released Amount (In Rs. Crore) 2020-21'].fillna(data['Released Amount (In Rs. Crore) 2020-21'].mean(), inplace=True)


data['Released Amount (In Rs. Crore) 2022-23'].fillna(data['Released Amount (In Rs. Crore) 2022-23'].mean(), inplace=True)


```

```python
data.isnull().sum()
```

```python
data.head()
```

#**Visualizations**

```python
# Violin Plot for Released Amount (In Rs. Crore) 2015-16 vs Physical Area (ha) 2018-19
figsize = (12, 1.2 * len(data['Physical Area (ha) 2018-19'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data=data, x='Released Amount (In Rs. Crore) 2015-16', y='Physical Area (ha) 2018-19', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.title('Physical Area (ha) 2018-19 vs Released Amount (In Rs. Crore) 2015-16')
plt.show()

# Violin Plot for Released Amount (In Rs. Crore) 2016-17 vs Physical Area (ha) 2019-20
figsize = (12, 1.2 * len(data['Physical Area (ha) 2019-20'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data=data, x='Released Amount (In Rs. Crore) 2016-17', y='Physical Area (ha) 2019-20', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.title('Physical Area (ha) 2019-20 vs Released Amount (In Rs. Crore) 2016-17')
plt.show()

# Violin Plot for Released Amount (In Rs. Crore) 2017-18 vs Physical Area (ha) 2020-21
figsize = (12, 1.2 * len(data['Physical Area (ha) 2020-21'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data=data, x='Released Amount (In Rs. Crore) 2017-18', y='Physical Area (ha) 2020-21', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.title('Physical Area (ha) 2020-21 vs Released Amount (In Rs. Crore) 2017-18')
plt.show()
```

```python
# Line Plot for Released Amount (In Rs. Crore) 2015-16
data['Released Amount (In Rs. Crore) 2015-16'].plot(kind='line', figsize=(8, 4), title='Released Amount (In Rs. Crore) 2015-16')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()

# Line Plot for Released Amount (In Rs. Crore) 2016-17
data['Released Amount (In Rs. Crore) 2016-17'].plot(kind='line', figsize=(8, 4), title='Released Amount (In Rs. Crore) 2016-17')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()



# Line Plot for Released Amount (In Rs. Crore) 2018-19
data['Released Amount (In Rs. Crore) 2018-19'].plot(kind='line', figsize=(8, 4), title='Released Amount (In Rs. Crore) 2018-19')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

```python
# Bar Plot for Physical Area (ha) 2018-19
data.groupby('Physical Area (ha) 2018-19').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title('Distribution of Physical Area (ha) 2018-19')
plt.show()

# Bar Plot for Physical Area (ha) 2019-20
data.groupby('Physical Area (ha) 2019-20').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title('Distribution of Physical Area (ha) 2019-20')
plt.show()

# Bar Plot for Physical Area (ha) 2020-21
data.groupby('Physical Area (ha) 2020-21').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title('Distribution of Physical Area (ha) 2020-21')
plt.show()

# Bar Plot for Physical Area (ha) 2021-22
data.groupby('Physical Area (ha) 2021-22').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title('Distribution of Physical Area (ha) 2021-22')
plt.show()
```

```python
# Histogram for Physical Area (ha) 2015-16
data['Physical Area (ha) 2015-16'].plot(kind='hist', bins=20, title='Physical Area (ha) 2015-16')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()

# Histogram for Physical Area (ha) 2016-17
data['Physical Area (ha) 2016-17'].plot(kind='hist', bins=20, title='Physical Area (ha) 2016-17')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()

# Histogram for Physical Area (ha) 2017-18
data['Physical Area (ha) 2017-18'].plot(kind='hist', bins=20, title='Physical Area (ha) 2017-18')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```
